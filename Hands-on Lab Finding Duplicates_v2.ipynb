## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
# Count and display duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows in the dataset: {duplicate_rows.shape[0]}")

# Display the first few duplicate rows to understand their structure
duplicate_rows.head()
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
# Identify duplicates based only on selected columns
subset_duplicates = df[df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork'])]

# Display number of such duplicate rows
print(f"Number of duplicate rows based on ['MainBranch', 'Employment', 'RemoteWork']: {subset_duplicates.shape[0]}")

# Show first few duplicate rows based on the subset
subset_duplicates.head()
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
import matplotlib.pyplot as plt

# Find full duplicate rows (not just subset)
duplicate_rows = df[df.duplicated(keep=False)]

# --- Plot 1: Distribution of duplicates by Country ---
plt.figure(figsize=(12, 6))
duplicate_rows['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries with Duplicate Entries")
plt.xlabel("Country")
plt.ylabel("Count of Duplicate Rows")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- Plot 2: Distribution of duplicates by Employment ---
plt.figure(figsize=(10, 6))
duplicate_rows['Employment'].value_counts().plot(kind='bar', color='orange')
plt.title("Distribution of Duplicates by Employment Type")
plt.xlabel("Employment Status")
plt.ylabel("Count of Duplicate Rows")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
# Define critical columns for uniqueness
unique_columns = ['MainBranch', 'Employment', 'Country', 'LanguageHaveWorkedWith']

# Remove duplicates based on critical subset columns
df_deduped = df.drop_duplicates(subset=unique_columns, keep='first')

# Display number of rows before and after
print(f"Original rows: {df.shape[0]}")
print(f"Rows after removing subset-based duplicates: {df_deduped.shape[0]}")
```
